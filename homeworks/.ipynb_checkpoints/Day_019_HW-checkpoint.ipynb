```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = 'D:/ML100Days/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train[~app_train["HOUSETYPE_MODE"].isna()]["HOUSETYPE_MODE"].unique()
app_train = app_train[~app_train["HOUSETYPE_MODE"].isna()]

nrows = 2
ncols = 2

plt.figure(figsize=(10,15))
for i in range(len(unique_house_type)):
    plt.subplot(3, 1, i+1)
    sns.distplot(app_train.loc[app_train["HOUSETYPE_MODE"] == unique_house_type[i] , ["AMT_CREDIT"]], kde=False)
    plt.title(str(unique_house_type[i]))
plt.show()
```
